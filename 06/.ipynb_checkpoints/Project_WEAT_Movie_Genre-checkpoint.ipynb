```python
import os
import pickle
import numpy as np


from konlpy.tag import Okt
from gensim.models import Word2Vec


import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


from numpy import dot   # for WEAT
from numpy.linalg import norm   # for WEAT
from sklearn.feature_extraction.text import TfidfVectorizer   #TF-IDF
```

```python
#     picke file   functions
def save_pkl(path, data):
    with open(path, "wb") as f:
        pickle.dump(data, f)

def load_pkl(path):
    with open(path, "rb") as f:
        data = pickle.load(f)
    return data

def read_token(file_path):
    okt = Okt()
    result = []
    with open(file_path, "r") as fread:
        print(file_path, 'Reading the file')
        while True:
            line = fread.readline()
            if not line: break
            tokenlist = okt.pos(line, stem=True, norm=True)
            for word in tokenlist:
                if word[1] in ["Noun"]:#, "Adjective", "Verb"]:
                    result.append((word[0]))
    return ' '.join(result)
```

```python
def cos_sim(i, j):
    return dot(i, j.T)/(norm(i)*norm(j))

def s(w, A, B):
    c_a = cos_sim(w, A)
    c_b = cos_sim(w, B)
    mean_A = np.mean(c_a, axis=-1)
    mean_B = np.mean(c_b, axis=-1)
    return mean_A - mean_B #, c_a, c_b

def weat_score(X, Y, A, B):

    s_X = s(X, A, B)
    s_Y = s(Y, A, B)

    mean_X = np.mean(s_X)
    mean_Y = np.mean(s_Y)

    std_dev = np.std(np.concatenate([s_X, s_Y], axis=0))

    return  (mean_X-mean_Y)/std_dev

```

```python

vectorizer = TfidfVectorizer()   #  
X = vectorizer.fit_transform([tokenized_art,tokenized_gen])   #TF-IDF 
print(X.shape)
```

```python
m1 = X[0].tocoo()   # art TF-IDF  sparse matrix 
m2 = X[1].tocoo()   # gen TF-IDF  sparse matrix 

w1 = [[i, j] for i, j in zip(m1.col, m1.data)]
w2 = [[i, j] for i, j in zip(m2.col, m2.data)]

w1.sort(key=lambda x: x[1], reverse=True)   # art   TF-IDF   
w2.sort(key=lambda x: x[1], reverse=True)   # gen   TF-IDF   
```

```python
print('  :')
for i in range(100):
    print(vectorizer.get_feature_names_out()[w1[i][0]], end=', ')

print('\n')

print('  :')
for i in range(100):
    print(vectorizer.get_feature_names_out()[w2[i][0]], end=', ')
```

```python
n = 15
w1_, w2_ = [], []
for i in range(100):
    w1_.append(vectorizer.get_feature_names_out()[w1[i][0]])
    w2_.append(vectorizer.get_feature_names_out()[w2[i][0]])

# w1  w2 ,     15 
target_art, target_gen = [], []
for i in range(100):
    if (w1_[i] not in w2_) and (w1_[i] in model.wv): target_art.append(w1_[i])
    if len(target_art) == n: break

# w2  w1 ,     15 
for i in range(100):
    if (w2_[i] not in w1_) and (w2_[i] in model.wv): target_gen.append(w2_[i])
    if len(target_gen) == n: break
```

```python
print('  : ', target_art)
print('  : ', target_gen)
```

```python
matrix = [[0 for _ in range(len(genre_name))] for _ in range(len(genre_name))]
print(len(matrix))
```

```python
X = np.array([model.wv[word] for word in target_art])
Y = np.array([model.wv[word] for word in target_gen])

for i in range(len(genre_name)-1):
    for j in range(i+1, len(genre_name)):
        A = np.array([model.wv[word] for word in attributes[i]])
        B = np.array([model.wv[word] for word in attributes[j]])
        matrix[i][j] = weat_score(X, Y, A, B)
```

- matrix . WEAT score  ,  .

```python
for i in range(len(genre_name)-1):
    for j in range(i+1, len(genre_name)):
        print(genre_name[i], genre_name[j], matrix[i][j])
```

```python
over8 = []
under_8 = []

for i in range(len(genre_name)-1):
    for j in range(i+1, len(genre_name)):
        if matrix[i][j] >= 0. :
            over8.append([i, j, matrix[i][j]])
        if matrix[i][j] <= -0.1 :
            under_8.append([i, j, matrix[i][j]])
```

```python
import pandas as pd
weat_score_df = pd.DataFrame({'X':['X'],
                             'Y':['Y'],
                             'A':['A'],
                             'B':['B'],
                             'WEAT Score':[0.0],
                             'Analysis':['Comment']})

for i in range(len(genre_name)-1):
    for j in range(i+1, len(genre_name)):
        if matrix[i][j] > 0.8:
          weat_score_df.loc[i] = ['', '', genre_name[i], genre_name[j], round(matrix[i][j], 2),
                                  f' {genre_name[i]}  ,  {genre_name[j]}   ']
        elif matrix[i][j] < -0.8:
          weat_score_df.loc[i] = ['', '', genre_name[i], genre_name[j], round(matrix[i][j], 2),
                                 f'{genre_name[i]}  ,  {genre_name[j]}   ']

weat_score_df
```

```python
from mecab import MeCab
```

```python

def read_token_mecab(file_path):
    mecab = MeCab()
    result = []
    with open(file_path, "r") as fread:
        print(file_path, 'Reading the file')
        while True:
            line = fread.readline()
            if not line: break
            tokenlist = mecab.pos(line)
            for word, pos in tokenlist:
                if pos.startswith("NNG"):  #   
                    result.append(word)
    return ' '.join(result)
```
